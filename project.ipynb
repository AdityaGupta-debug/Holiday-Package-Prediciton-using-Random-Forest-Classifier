```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['TypeofContact'] = df['TypeofContact'].fillna(df['TypeofContact'].mode()[0])
```

```python
df['NumberOfFollowups'] = df['NumberOfFollowups'].fillna(df['NumberOfFollowups'].mean())
```

```python
df['PreferredPropertyStar'] = df['PreferredPropertyStar'].fillna(df['PreferredPropertyStar'].mode()[0])
```

```python
df['NumberOfTrips'] = df['NumberOfTrips'].fillna(df['NumberOfTrips'].mean())
```

```python
df['NumberOfChildrenVisiting'] = df['NumberOfChildrenVisiting'].fillna(df['NumberOfChildrenVisiting'].mode()[0])
```

```python
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())
```

```python
df['DurationOfPitch'] = df['DurationOfPitch'].fillna(df['DurationOfPitch'].median())
```

```python
df = df.drop('CustomerID',axis=1)
df.head()
```

```python
df['Total_visiting']  = df['NumberOfChildrenVisiting'] + df['NumberOfPersonVisiting']
```

```python
df =  df.drop(['NumberOfChildrenVisiting', 'NumberOfPersonVisiting'], axis=1)
```

```python
cat_features = [feature for feature in df.columns if df[feature].dtype == 'O']
cat_features
```

```python
num_features = [feature for feature in df.columns if df[feature].dtype != 'O']
num_features
```

```python
X = df.iloc[:, 1:]
X
```

```python
Y = df.iloc[:, 0]
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
cat_features = X.select_dtypes(include="object").columns
num_features = X.select_dtypes(exclude="object").columns

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

transformer = ColumnTransformer([
    ('trf1', OneHotEncoder(drop='first'),cat_features),
    ('trf2', StandardScaler(), num_features) 
], remainder='passthrough')
```

```python
transformer
```

```python
X_train= transformer.fit_transform(X_train)
```

```python
pd.DataFrame(X_train)
```

```python
X_test = transformer.transform(X_test)
```

```python
pd.DataFrame(X_test)
```

```python
Y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model1 = RandomForestClassifier()
model1.fit(X_train, Y_train)
y_pred1 = model1.predict(X_test)
score1 = accuracy_score(Y_test, y_pred1)
print(score1)
```

```python
model2 = LogisticRegression()
model2.fit(X_train, Y_train)
y_pred2 = model2.predict(X_test)
score2 = accuracy_score(Y_test, y_pred2)
print(score2)
```

```python
model3 = DecisionTreeClassifier()
model3.fit(X_train, Y_train)
y_pred3 = model3.predict(X_test)
score3 = accuracy_score(Y_test, y_pred3)
print(score3)
```

```python
model4= AdaBoostClassifier()
model4.fit(X_train, Y_train)
y_pred4 = model4.predict(X_test)
score3 = accuracy_score(Y_test, y_pred4)
print(score3)
```
